```python
import matplotlib.pyplot as plt
import pandas as pd
```

## Read CSV File

```python
df = pd.read_csv('dummy.csv')
df.head()
```

## Bar in Matplotlib

```python
plt.bar(df.Name, df.Percentage, color='green')
plt.show()
```

## Pie - Chart in Matplotlib

```python
plt.pie(df.Percentage, labels=df.Name)
plt.show()
```

## Ploting in Matplotlib

```python
plt.plot(df.Name, df.Percentage, color='red', marker='o')
plt.grid()
plt.show()
```

## Scatter in Matplotlib

```python
plt.scatter(df.Name, df.Percentage)
plt.show()
```
